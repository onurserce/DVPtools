```python
import os
import skimage
import matplotlib.pyplot as plt
from math import sqrt
import numpy as np
from aicsimageio import AICSImage
```

```python
data = AICSImage(r"C:\Users\onur_serce\Desktop\komal_stacked_for_segmentation\mip_T0.ome.tiff")
```

```python
data.dims
```

```python
ch1 = data.get_image_data("YX", T=0, C=1, z=0)
```

```python
import matplotlib.patches as patches

plt.imshow(ch1)
ax = plt.gca()
rect = patches.Rectangle((95, 50), 50, 80, linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rect)
```

```python
from skimage.feature import blob_log

blobs = blob_log(ch1[50:130, 95:155], min_sigma=1, max_sigma=2, num_sigma=20, threshold=0.25, overlap=0.5, log_scale=False, threshold_rel=None, exclude_border=False)

blobs[:, 2] = blobs[:, 2] * sqrt(2)

plt.imshow(ch1[50:130, 95:155])
ax = plt.gca()
for blob in blobs:
    y, x, r = blob
    c = plt.Circle((x, y), r, color='red', linewidth=2, fill=False)
    ax.add_patch(c)

plt.tight_layout()
```
